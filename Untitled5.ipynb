# Static visualizations in Matplotlib are non-interactive plots commonly used to represent data in a fixed form. These visualizations are perfect for reports, presentations, or any scenario where user interaction is not required. It includes line plot, bar plot etc

```python
import matplotlib.pyplot as plt

# Data
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create the plot
plt.plot(x, y, label='y = x^2', color='blue', marker='o')

# Add labels, title, and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()

# Show the plot
plt.show()
```

# Animated visualizations in Matplotlib allow you to create dynamic visual representations of data that change over time. These can be useful for illustrating concepts, time-series data, or other scenarios where movement helps convey information.Matplotlib provides the matplotlib.animation module for creating animations.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

# Create the figure
fig, ax = plt.subplots()
bars = plt.bar(categories, values)

# Update function
def update(frame):
    for bar in bars:
        bar.set_height(np.random.randint(1, 10))  # Random heights
    return bars

# Create the animation
ani = FuncAnimation(fig, update, frames=50, interval=200, blit=True)

# Save the animation as a GIF
ani.save('bar_animation.gif', writer='pillow')

# Display the animation in Colab
from IPython.display import Image
Image(open('bar_animation.gif', 'rb').read())
```

# Matplotlib supports interactive visualizations, enabling users to explore data by zooming, panning, or interacting with widgets like sliders, buttons, and text boxes. Interactive visualizations are especially useful for exploratory data analysis.

# To display interactive 3D plots in Google Colab, you need to use the plotly or pythreejs backends because the default Matplotlib 3D plots are not fully interactive in Colab. Below is the code for running an interactive 3D scatter plot in Colab using Plotly.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

ax.scatter(x, y, z, c=z, cmap='viridis', s=50)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.title('Interactive 3D Scatter Plot')
plt.show()
```

```python
import plotly.express as px
import numpy as np

# Generate random data
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Create 3D scatter plot
fig = px.scatter_3d(
    x=x,
    y=y,
    z=z,
    color=z,  # Use z values for color
    labels={'x': 'X-axis', 'y': 'Y-axis', 'z': 'Z-axis'},
    title="Interactive 3D Scatter Plot"
)

# Show the plot
fig.show()
```
